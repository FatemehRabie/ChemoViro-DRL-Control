```python
import numpy as np
import gymnasium as gym
from scipy.integrate import solve_ivp
from scipy.interpolate import interpn
from stable_baselines3 import TD3, SAC, PPO
from sb3_contrib import TRPO
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.callbacks import BaseCallback
import time
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import optuna

# ===================================================================
# Configuration Section
# ===================================================================

CONFIG = {
    "ode_params": {
        "default": {
            "alpha": 0.206, "K": 1e6, "beta": 0.05, "Ku": 1e5, "delta_U": 50.0,
            "Kc": 1e5, "delta": 0.5115, "delta_I": 60.0, "b": 50,
            "gamma": 0.01, "psi": 4.17, "nu_U": 0.08, "nu_I": 0.1,
            "tau": 0.2, "delta_V": 0.01, "delta_T": 0.01, "phi": 0.7,
            "beta_T": 0.5, "kappa": 1e5
        },
        "modified": {
            "alpha": 0.5, "K": 1e8, "beta": 0.1, "Ku": 1000.0, "delta_U": 25.0,         #yadam bashe  k ro az 10^6 be 10^8 taghir dadam
            "Kc": 1e7, "delta": 0.7117, "delta_I": 70.0, "b": 100,
            "gamma": 0.03, "psi": 7.17, "nu_U": 0.08, "nu_I": 0.1, "tau": 0.2,
            "delta_V": 0.01, "delta_T": 0.01, "phi": 0.7, "beta_T": 0.5,
            "kappa": 1e5
        },

        "aggressive": {
            "alpha": 0.55,      #       
            "K": 5e6,           #        
            "beta": 0.08,       #      
            "Ku": 1000.0,       
            "delta_U": 45.0,    
            "Kc": 1.5e5,        
            "delta": 0.65,      #     
            "delta_I": 75.0,    #      
            "b": 120,           #    
            "gamma": 0.02,      
            "psi": 9.0,         #         
            "nu_U": 0.08, "nu_I": 0.1, "tau": 0.2,
            "delta_V": 0.01, "delta_T": 0.01, "phi": 0.7, "beta_T": 0.5,
            "kappa": 1e5
        }

    },
    "noise_config": {
        "none": {"initial_conditions_noise": 0.0, "observations_noise": 0.0},
        "default": {"initial_conditions_noise": 0.05, "observations_noise": 0.01}
    },
    "rl_algorithms": ["TRPO", "TD3", "SAC", "PPO"],
    "training_steps": 100000,               # total training az 50000 ro be 100000 taghir dadam 
    "tuning_trials": 15,              # tedad emtehanat baraye tuning hyperparameter ha az 10 be 15 taghir dadam
    "reward_weights": {"A1": 1e-7, "A2": 1e-5}       # inam ghablan A1=1e-5 va A2=1e-3 bood

}

# ===================================================================
# ODE System Definitions
# ===================================================================

def reduced_chemo_viro_rhs(t, y, action, params):
    U, I, V, C = y; u1, u2 = action
    U, I, V, C = max(0, U), max(0, I), max(0, V), max(0, C)
    dUdt = (params["alpha"]*U*(1 - (U+I)/params["K"]) - (params["beta"]*U*V)/(params["Ku"]+U) - (params["delta_U"]*U*C)/(params["Kc"]+C))
    dIdt = ((params["beta"]*U*V)/(params["Ku"]+U) - params["delta"]*I - (params["delta_I"]*I*C)/(params["Kc"]+C))
    dVdt = params["b"]*params["delta"]*I - (params["beta"]*U*V)/(params["Ku"]+U) - params["gamma"]*V + u1
    dCdt = u2 - params["psi"]*C
    return np.array([dUdt, dIdt, dVdt, dCdt], dtype=np.float32)

def full_chemo_viro_rhs(t, y, action, params):
    U, I, V, E_V, E_T, C = y; u1, u2 = action
    U,I,V,E_V,E_T,C=max(0,U),max(0,I),max(0,V),max(0,E_V),max(0,E_T),max(0,C)
    dUdt = (params["alpha"]*U*(1-(U+I)/params["K"])-(params["beta"]*U*V)/(params["Ku"]+U)-params["nu_U"]*U*E_T-(params["delta_U"]*U*C)/(params["Kc"]+C))
    dIdt = ((params["beta"]*U*V)/(params["Ku"]+U)-params["delta"]*I-params["nu_I"]*E_T*I-params["tau"]*E_V*I-(params["delta_I"]*I*C)/(params["Kc"]+C))
    dVdt = params["b"]*params["delta"]*I - (params["beta"]*U*V)/(params["Ku"]+U) - params["gamma"]*V+u1
    dE_Vdt = params["phi"]*I - params["delta_V"]*E_V
    dE_Tdt = params["beta_T"]*(U+I)/(params["kappa"]+U+I) - params["delta_T"]*E_T
    dCdt = u2 - params["psi"]*C
    return np.array([dUdt, dIdt, dVdt, dE_Vdt, dE_Tdt, dCdt], dtype=np.float32)

# ===================================================================
# Gymnasium Environment
# ===================================================================

class ChemoViroEnv(gym.Env):
    def __init__(self, ode_model='reduced', ode_params=None, noise_config=None, max_time=400.0, dt=1.0, reward_weights=None):
        super().__init__()
        self.ode_model, self.ode_params, self.noise_config, self.max_time, self.dt, self.reward_weights = ode_model, ode_params or {}, noise_config or {}, max_time, dt, reward_weights or {}
        if self.ode_model == 'reduced':
            self.ode_rhs, self.state_dim, high_obs = reduced_chemo_viro_rhs, 4, np.array([1e6, 1e6, 1e5, 1e5], dtype=np.float32)
        elif self.ode_model == 'full':
            self.ode_rhs, self.state_dim, high_obs = full_chemo_viro_rhs, 6, np.array([1e6, 1e6, 1e5, 1e5, 1e5, 1e5], dtype=np.float32)
        self.observation_space = gym.spaces.Box(low=np.zeros(self.state_dim, dtype=np.float32), high=high_obs, dtype=np.float32)
        self.action_space = gym.spaces.Box(low=np.array([0, 0]), high=np.array([1000, 100]), dtype=np.float32)

    def reset(self, seed=None, options=None):
        super().reset(seed=seed); self.time = 0.0
        base_state = [10000, 100, 500, 100] if self.ode_model == 'reduced' else [10000, 100, 500, 100, 100, 100]
        noise = self.noise_config.get("initial_conditions_noise", 0)
        self.state = np.array([s * (1 + noise * np.random.normal()) for s in base_state], dtype=np.float32)
        return self._get_obs(), {}

    def step(self, action):
        action = np.clip(action, self.action_space.low, self.action_space.high)
        sol = solve_ivp(lambda t, y: self.ode_rhs(t, y, action, self.ode_params), (self.time, self.time + self.dt), self.state, method='Radau', rtol=1e-5, atol=1e-8)
        self.state = sol.y[:, -1].astype(np.float32)
        if np.any(np.isnan(self.state)) or np.any(np.isinf(self.state)):
            return self._get_obs(), -1e6, True, False, {}
        self.time += self.dt; self.state = np.clip(self.state, 0, 1e8)
        U, I = self.state[0], self.state[1]; current_tumor_size = U + I
        reward = -current_tumor_size - (0.5*self.reward_weights["A1"]*action[0]**2 + 0.5*self.reward_weights["A2"]*action[1]**2)
        terminated = self.time >= self.max_time or current_tumor_size < 0.1
        return self._get_obs(), reward, terminated, False, {}

    def _get_obs(self):
        obs = self.state.copy()
        if np.any(np.isnan(obs)) or np.any(np.isinf(obs)):
            obs = np.nan_to_num(obs, nan=0.0, posinf=1e8, neginf=0.0)
        noise = self.noise_config.get("observations_noise", 0)
        if noise > 0: obs *= (1 + noise * np.random.normal(size=obs.shape))
        return obs

# ===================================================================
# Callbacks, Hyperparameter Tuning, and Training
# ===================================================================

class RewardCallback(BaseCallback):
    def __init__(self, check_freq: int, verbose=0):
        super(RewardCallback, self).__init__(verbose)
        self.check_freq = check_freq
        self.rewards = []
        self.timesteps = []

    def _on_step(self) -> bool:
        if self.n_calls % self.check_freq == 0:
            if hasattr(self.model, 'ep_info_buffer') and len(self.model.ep_info_buffer) > 0:
                mean_reward = np.mean([ep_info['r'] for ep_info in self.model.ep_info_buffer])
                self.rewards.append(mean_reward)
                self.timesteps.append(self.num_timesteps)
        return True

def define_search_space(trial, algo):
    lr = trial.suggest_float("lr", 1e-5, 1e-3, log=True)
    gamma = trial.suggest_float("gamma", 0.9, 0.9999)
    net_arch_size = trial.suggest_categorical("net_arch_size", [128, 256, 512])

    if algo in ["TD3", "SAC"]:
        policy_kwargs = dict(net_arch=dict(pi=[net_arch_size, net_arch_size], qf=[net_arch_size, net_arch_size]))
    elif algo in ["TRPO", "PPO"]:
        policy_kwargs = dict(net_arch=dict(pi=[net_arch_size, net_arch_size], vf=[net_arch_size, net_arch_size]))
    else:
        policy_kwargs = {}

    params = {"lr": lr, "gamma": gamma, "policy_kwargs": policy_kwargs}

    if algo == "PPO":
        params["n_steps"] = trial.suggest_categorical("n_steps", [256, 512, 1024])
        params["batch_size"] = trial.suggest_categorical("batch_size", [64, 128])
        params["n_epochs"] = trial.suggest_categorical("n_epochs", [5, 10])
        params["clip_range"] = trial.suggest_float("clip_range", 0.1, 0.4)

    return params

def objective(trial, algo, env):
    try:
        hyperparams = define_search_space(trial, algo)

        common_kwargs = dict(env=env, verbose=0, seed=42, gamma=hyperparams["gamma"], learning_rate=hyperparams["lr"], policy_kwargs=hyperparams["policy_kwargs"].copy())

        if algo == "TRPO":
            model = TRPO('MlpPolicy', **common_kwargs)
        elif algo == "TD3":
            model = TD3('MlpPolicy', **common_kwargs)
        elif algo == "SAC":
            model = SAC('MlpPolicy', **common_kwargs)
        elif algo == "PPO":
            model_kwargs = {**common_kwargs,
                            "n_steps": hyperparams["n_steps"],
                            "batch_size": hyperparams["batch_size"],
                            "n_epochs": hyperparams["n_epochs"],
                            "clip_range": hyperparams["clip_range"]}
            model = PPO('MlpPolicy', **model_kwargs)
        else:
             raise ValueError(f"Algorithm {algo} not supported")

        model.learn(total_timesteps=CONFIG["training_steps"] // 4)
        reward = run_simulation(env, model)
        return reward
    except (AssertionError, ValueError) as e:
        print(f"Trial pruned due to error: {e}")
        raise optuna.exceptions.TrialPruned()

def run_simulation(env, model, return_actions=False):
    obs, _ = env.reset(seed=42)
    states_history, rewards_history, actions_history = [], [], []
    done = False
    while not done:
        if isinstance(model, (DPPolicy, ConstantPolicy)):
             action, _ = model.predict(obs, deterministic=True)
        else:
            model_obs_shape = model.observation_space.shape[0]
            env_obs_shape = env.observation_space.shape[0]
            
            if model_obs_shape != env_obs_shape:
                if model_obs_shape == 6 and env_obs_shape == 4:
                    padded_obs = np.insert(obs, 3, [0, 0])
                    action, _ = model.predict(padded_obs, deterministic=True)
                elif model_obs_shape == 4 and env_obs_shape == 6:
                    reduced_obs = np.delete(obs, [3, 4])
                    action, _ = model.predict(reduced_obs, deterministic=True)
                else:
                    raise ValueError(f"Unhandled mismatch: Model expects {model_obs_shape}D, Env provides {env_obs_shape}D")
            else:
                action, _ = model.predict(obs, deterministic=True)

        obs, reward, done, _, _ = env.step(action)
        states_history.append(obs)
        rewards_history.append(reward)
        actions_history.append(action)

    if return_actions:
        return np.sum(rewards_history), np.array(actions_history)
    return np.sum(rewards_history)

# ===================================================================
# Baseline and Dynamic Programming Policies
# ===================================================================

class ConstantPolicy:
    def __init__(self, action_value):
        self.action_value = np.array(action_value)
        self.observation_space = gym.spaces.Box(low=-np.inf, high=np.inf, shape=(1,))
    def predict(self, observation, deterministic=True):
        return self.action_value, None

class DPPolicy:
    def __init__(self, policy_tables, state_grids, actions_grid):
        self.policy_u1, self.policy_u2, self.state_grids, self.actions_grid = policy_tables[0], policy_tables[1], state_grids, actions_grid
        self.observation_space = gym.spaces.Box(low=0, high=1, shape=(4,))
    def predict(self, observation, deterministic=True):
        u1 = interpn(self.state_grids, self.policy_u1[0, ...], observation, bounds_error=False, fill_value=None)[0]
        u2 = interpn(self.state_grids, self.policy_u2[0, ...], observation, bounds_error=False, fill_value=None)[0]
        return self.actions_grid[np.argmin(np.linalg.norm(self.actions_grid - np.array([u1, u2]), axis=1))], None

def run_dynamic_programming(params):
    print("--- Calculating Dynamic Programming Policy ---")
    start_time = time.time()
    num_points = {'U': 10, 'I': 12, 'V': 10, 'C': 10} 
    state_grids = (np.linspace(0,1e6,num_points['U']), np.linspace(0,1e6,num_points['I']), np.linspace(0,1e5,num_points['V']), np.linspace(0,50,num_points['C']))
    actions = np.array(np.meshgrid([0,100,200], [0,25,50])).T.reshape(-1, 2)
    T_final, dt = 100, 5
    N = len(np.arange(0, T_final + dt, dt))
    value_table = np.zeros((N, num_points['U'], num_points['I'], num_points['V'], num_points['C']))
    p_u1, p_u2 = np.zeros_like(value_table), np.zeros_like(value_table)
    for k in range(N - 2, -1, -1):
        for iU, U in enumerate(state_grids[0]):
            for iI, I in enumerate(state_grids[1]):
                for iV, V in enumerate(state_grids[2]):
                    for iC, C in enumerate(state_grids[3]):
                        min_cost, best_action = np.inf, actions[0]
                        for action in actions:
                            cost = (U+I + (CONFIG["reward_weights"]["A1"]/2)*action[0]**2 + (CONFIG["reward_weights"]["A2"]/2)*action[1]**2) * dt
                            sol = solve_ivp(lambda t,y: reduced_chemo_viro_rhs(t,y,action,params), [0,dt], [U,I,V,C], dense_output=True)
                            future_cost = interpn(state_grids, value_table[k+1,...], sol.y[:,-1], bounds_error=False, fill_value=np.inf)[0]
                            if cost + future_cost < min_cost:
                                min_cost, best_action = cost + future_cost, action
                        value_table[k,iU,iI,iV,iC] = min_cost
                        p_u1[k,iU,iI,iV,iC], p_u2[k,iU,iI,iV,iC] = best_action
    runtime = time.time() - start_time
    print(f"--- DP calculation finished in {runtime:.2f} seconds ---")
    return DPPolicy((p_u1, p_u2), state_grids, actions), runtime

# ===================================================================
# Plotting and Reporting
# ===================================================================
plt.rcParams.update({
    'font.size': 14,         
    'axes.titlesize': 18,     
    'axes.labelsize': 16,    
    'xtick.labelsize': 14,    
    'ytick.labelsize': 14,    
    'legend.fontsize': 16,    
    'figure.titlesize': 20    
})

def plot_training_rewards(all_reward_histories):
    sns.set_style("whitegrid") 
    num_scenarios = len(all_reward_histories)
    if num_scenarios == 0: return
    fig, axes = plt.subplots(num_scenarios, 1, figsize=(12, 5 * num_scenarios), sharex=True)
    if num_scenarios == 1: axes = [axes]
    for ax, (scenario_name, histories) in zip(axes, all_reward_histories.items()):
        for algo, (timesteps, rewards) in histories.items():
            ax.plot(timesteps, rewards, label=algo)
        ax.set_title(f"Training Progress: {scenario_name}")
        ax.set_ylabel("Mean Reward per Episode")
        ax.legend(); ax.grid(True)
    axes[-1].set_xlabel("Timesteps")
    plt.tight_layout()
    plt.savefig("training_rewards.png")
    plt.show()

def plot_successful_actions(all_results, baseline_rewards):
    sns.set_style("darkgrid")
    print("\n\n" + "="*80 + "\n" + " " * 20 + "PLOTTING ACTIONS FOR SUCCESSFUL POLICIES\n" + "="*80)
    for result in all_results:
        if result['name'].startswith("Model 2"):
            continue
        for test_name, (reward, actions) in result['test_results'].items():
            if test_name in baseline_rewards and not np.isnan(reward) and reward > baseline_rewards[test_name]: # Added check for nan
                print(f"Plotting actions for '{result['name']}' on '{test_name}' (Reward: {reward:.2e} > Baseline: {baseline_rewards[test_name]:.2e})")
                plt.figure(figsize=(12, 5))
                plt.plot(actions[:, 0], label='Virus Dose (u1)'); plt.plot(actions[:, 1], label='Drug Dose (u2)')
                plt.title(f"Control Actions: {result['name']} on {test_name}"); plt.xlabel("Time Step"); plt.ylabel("Dose")
                plt.legend(); plt.grid(True)

                filename = f"actions_{result['name']}_{test_name}.png".replace(" ", "_").replace("(", "").replace(")", "").replace("/", "")
                plt.savefig(filename)
                print(f"Saved {filename}")
                plt.show()


def plot_model2_actions(all_results):
    sns.set_style("darkgrid")

    print("\n\n" + "="*80 + "\n" + " " * 20 + "PLOTTING ACTIONS FOR 'MODEL 2' (ALL ALGORITHMS)\n" + "="*80)
    for result in all_results:
        if result['name'].startswith("Model 2"):
            for test_name, (reward, actions) in result['test_results'].items():
                if not np.isnan(reward): # Added check for nan
                    print(f"Plotting actions for '{result['name']}' on '{test_name}' (Reward: {reward:.2e})")
                    plt.figure(figsize=(12, 5))
                    plt.plot(actions[:, 0], label='Virus Dose (u1)')
                    plt.plot(actions[:, 1], label='Drug Dose (u2)')
                    plt.title(f"Control Actions: {result['name']} on {test_name}")
                    plt.xlabel("Time Step")
                    plt.ylabel("Dose")
                    plt.legend()
                    plt.grid(True)
                    filename = f"actions_{result['name']}_{test_name}.png".replace(" ", "_").replace("(", "").replace(")", "").replace("/", "")
                    plt.savefig(filename)
                    print(f"Saved {filename}")
                    plt.show()

# ===================================================================
# Main Execution Block
# ===================================================================

def main():
    training_scenarios = [
        {"name": "Model 1 (Full, Default, Noise)", "ode_model": "full", "params": "default", "noise": "default"},
        {"name": "Model 2 (Reduced, Default, Noise)", "ode_model": "reduced", "params": "default", "noise": "default"},
        {"name": "Model 3 (Reduced, Modified, Noise)", "ode_model": "reduced", "params": "modified", "noise": "default"},
        {"name": "Model 4 (Reduced, Modified, No Noise)", "ode_model": "reduced", "params": "modified", "noise": "none"},
        {"name": "Model 5 (Reduced, Aggressive, Noise)", "ode_model": "reduced", "params": "aggressive", "noise": "default"},
    ]
    trained_models, runtimes, all_reward_histories = {}, [], {}
    for scenario in training_scenarios:
        print("\n" + "="*60 + f"\nTRAINING SCENARIO: {scenario['name']}\n" + "="*60)
        env = ChemoViroEnv(
            ode_model=scenario["ode_model"], ode_params=CONFIG["ode_params"][scenario["params"]],
            noise_config=CONFIG["noise_config"][scenario["noise"]], reward_weights=CONFIG["reward_weights"]
        )
        trained_models[scenario['name']], scenario_histories = {}, {}
        for algo in CONFIG["rl_algorithms"]:
            print(f"\n--- Tuning {algo} for {scenario['name']} ---")
            study = optuna.create_study(direction="maximize")
            study.optimize(lambda trial: objective(trial, algo, Monitor(env)), n_trials=CONFIG["tuning_trials"])
            print(f"Best trial for {algo}: {study.best_trial.value:.2e} with params {study.best_trial.params}")

            best_hyperparams = define_search_space(study.best_trial, algo)

            print(f"--- Retraining best {algo} model ---")

            common_kwargs = dict(env=Monitor(env), verbose=0, seed=42, gamma=best_hyperparams["gamma"], learning_rate=best_hyperparams["lr"])
            if algo in ["TRPO", "TD3", "SAC"]:
                model_kwargs = {**common_kwargs, "policy_kwargs": best_hyperparams["policy_kwargs"].copy()}
                if algo == "TRPO": model = TRPO('MlpPolicy', **model_kwargs)
                elif algo == "TD3": model = TD3('MlpPolicy', **model_kwargs)
                elif algo == "SAC": model = SAC('MlpPolicy', **model_kwargs)
            elif algo == "PPO":
                model_kwargs = {**common_kwargs,
                                "policy_kwargs": best_hyperparams["policy_kwargs"].copy(),
                                "n_steps": best_hyperparams["n_steps"],
                                "batch_size": best_hyperparams["batch_size"],
                                "n_epochs": best_hyperparams["n_epochs"],
                                "clip_range": best_hyperparams["clip_range"]}
                model = PPO('MlpPolicy', **model_kwargs)

            reward_callback = RewardCallback(check_freq=100)
            start_time = time.time()
            model.learn(total_timesteps=CONFIG["training_steps"], callback=reward_callback)
            runtime = time.time() - start_time

            trained_models[scenario['name']][algo] = model
            scenario_histories[algo] = (reward_callback.timesteps, reward_callback.rewards)
            runtimes.append({"Scenario/Model": scenario['name'], "Algorithm": f"{algo} (Tuned)", "Runtime (s)": f"{runtime:.2f}"})
        all_reward_histories[scenario['name']] = scenario_histories

    dp_policy, dp_runtime = run_dynamic_programming(CONFIG["ode_params"]["default"])
    runtimes.append({"Scenario/Model": "Dynamic Programming", "Algorithm": "Policy Calculation", "Runtime (s)": f"{dp_runtime:.2f}"})

    testing_scenarios = [
        {"name": "Test 1 (Reduced, Default, Noise)", "ode_model": "reduced", "params": "default", "noise": "default"},
        {"name": "Test 2 (Reduced, Default, No Noise)", "ode_model": "reduced", "params": "default", "noise": "none"},
        {"name": "Test 3 (Full, Modified, Noise)", "ode_model": "full", "params": "modified", "noise": "default"},
        {"name": "Test 4 (Reduced, Aggressive, Noise)", "ode_model": "reduced", "params": "aggressive", "noise": "default"},
    ]

    all_results = []
    for train_scenario_name, models in trained_models.items():
        for algo, model in models.items():
            result_record = {"name": f"{train_scenario_name} - {algo}", "test_results": {}}
            for test_scenario in testing_scenarios:
                test_env = ChemoViroEnv(ode_model=test_scenario["ode_model"], ode_params=CONFIG["ode_params"][test_scenario["params"]], noise_config=CONFIG["noise_config"][test_scenario["noise"]], reward_weights=CONFIG["reward_weights"])
                reward, actions = run_simulation(test_env, model, return_actions=True)
                result_record["test_results"][test_scenario['name']] = (reward, actions)
            all_results.append(result_record)

    dp_result_record = {"name": "Dynamic Programming Baseline", "test_results": {}}
    for test_scenario in testing_scenarios:
        test_env = ChemoViroEnv(ode_model=test_scenario["ode_model"], ode_params=CONFIG["ode_params"][test_scenario["params"]], noise_config=CONFIG["noise_config"][test_scenario["noise"]], reward_weights=CONFIG["reward_weights"])
        if test_env.ode_model == 'reduced':
            reward, actions = run_simulation(test_env, dp_policy, return_actions=True)
            dp_result_record["test_results"][test_scenario['name']] = (reward, actions)
        else:
            dp_result_record["test_results"][test_scenario['name']] = (np.nan, np.nan) 
    all_results.append(dp_result_record)

    print("--- Evaluating Constant Policy Baseline ---")
    start_time = time.time()

    constant_policy = ConstantPolicy(action_value=[500, 50])
    constant_result_record = {"name": "Constant Policy Baseline", "test_results": {}}
    for test_scenario in testing_scenarios:
        test_env = ChemoViroEnv(ode_model=test_scenario["ode_model"], ode_params=CONFIG["ode_params"][test_scenario["params"]], noise_config=CONFIG["noise_config"][test_scenario["noise"]], reward_weights=CONFIG["reward_weights"])
        reward, actions = run_simulation(test_env, constant_policy, return_actions=True)
        constant_result_record["test_results"][test_scenario['name']] = (reward, actions)
    all_results.append(constant_result_record)

    runtime = time.time() - start_time
    runtimes.append({"Scenario/Model": "Constant Policy Baseline", "Algorithm": "Policy Evaluation", "Runtime (s)": f"{runtime:.2f}"})
    print(f"--- Constant Policy evaluation finished in {runtime:.2f} seconds ---")
    table_results = []
    for result in all_results:
        record = {"Trained Model": result["name"]}
        for test_name, (reward, *_) in result["test_results"].items():
              record[test_name] = f"{reward:.2e}"
        table_results.append(record)

    print("\n\n" + "="*60 + "\n" + " " * 15 + "COMPUTATION TIMES\n" + "="*60)
    print(pd.DataFrame(runtimes).to_string(index=False))

    print("\n\n" + "="*80 + "\n" + " " * 20 + "FINAL PERFORMANCE: CUMULATIVE REWARDS\n" + "="*80)
    print(pd.DataFrame(table_results).to_string(index=False))
    print("="*80)

    plot_training_rewards(all_reward_histories)

    baseline_rewards = {test_name: reward for test_name, (reward, _) in constant_result_record["test_results"].items()}
    plot_successful_actions(all_results, baseline_rewards)

    plot_model2_actions(all_results)

    return all_results, all_reward_histories, runtimes, trained_models


if __name__ == "__main__":
    results_data, rewards_data, runtimes_data, models_data = main()

    print("\n" + "="*50)
    print("Execution finished.")
    print("You can now access the following variables for further analysis:")
    print("- results_data: Contains cumulative rewards and action histories for all tests.")
    print("- rewards_data: Contains training reward histories.")
    print("- runtimes_data: Contains computation times.")
    print("- models_data: Contains the trained SB3 model objects.")
    print("="*50 + "\n")
```
